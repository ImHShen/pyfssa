# 使用print语句调试
通过在代码中插入print语句来输出变量值和程序状态。

```python
# 使用print语句调试

# 示例代码
def add_numbers(a, b):
    result = a + b
    print(f"add_numbers: a={a}, b={b}, result={result}")  # 输出变量值
    return result

# 调用函数并查看输出
sum_result = add_numbers(5, 7)
print(f"sum_result={sum_result}")  # 输出最终结果

# 另一个示例
def find_max(numbers):
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
        print(f"find_max: num={num}, max_num={max_num}")  # 输出循环中的变量值
    return max_num

# 调用函数并查看输出
max_result = find_max([3, 1, 4, 1, 5, 9, 2, 6, 5])
print(f"max_result={max_result}")  # 输出最终结果
```

# 使用断点调试
在代码中插入断点，并使用调试工具进行逐步调试。

```python
# 使用断点调试

# 示例代码
def add_numbers(a, b):
    result = a + b
    pdb.set_trace()  # 设置断点
    print(f"add_numbers: a={a}, b={b}, result={result}")  # 输出变量值
    return result

# 调用函数并查看输出
sum_result = add_numbers(5, 7)
print(f"sum_result={sum_result}")  # 输出最终结果

# 另一个示例
def find_max(numbers):
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
        pdb.set_trace()  # 设置断点
        print(f"find_max: num={num}, max_num={max_num}")  # 输出循环中的变量值
    return max_num

# 调用函数并查看输出
max_result = find_max([3, 1, 4, 1, 5, 9, 2, 6, 5])
print(f"max_result={max_result}")  # 输出最终结果
```

# 使用%debug魔法命令
在发生异常后使用%debug魔法命令进入调试模式。

```python
# 使用%debug魔法命令

# 示例代码
def divide_numbers(a, b):
    result = a / b  # 可能会引发ZeroDivisionError异常
    print(f"divide_numbers: a={a}, b={b}, result={result}")  # 输出变量值
    return result

# 调用函数并查看输出
try:
    divide_result = divide_numbers(10, 0)
except ZeroDivisionError:
    %debug  # 进入调试模式

# 另一个示例
def get_item_from_list(lst, index):
    item = lst[index]  # 可能会引发IndexError异常
    print(f"get_item_from_list: lst={lst}, index={index}, item={item}")  # 输出变量值
    return item

# 调用函数并查看输出
try:
    item_result = get_item_from_list([1, 2, 3], 5)
except IndexError:
    %debug  # 进入调试模式
```
